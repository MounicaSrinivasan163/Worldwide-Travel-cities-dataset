```python
import json

# Function to parse and flatten the avg_temp_monthly JSON string
def parse_monthly_temp(json_str):
    try:
        temp_dict = json.loads(json_str)
        flat = {}
        for month, stats in temp_dict.items():
            for stat, value in stats.items():
                flat[f"{stat}_{month}"] = value
        return pd.Series(flat)
    except:
        return pd.Series()

# Apply parsing function to the avg_temp_monthly column
monthly_temp_expanded = df['avg_temp_monthly'].apply(parse_monthly_temp)

# Create new DataFrame with city + the 36 temp columns
df_temp = pd.concat([df['city'], monthly_temp_expanded], axis=1)

# (Optional) Reorder columns: city, avg_1...avg_12, max_1...max_12, min_1...min_12
months = [str(i) for i in range(1, 13)]
ordered_cols = (
    ['city'] +
    [f"avg_{m}" for m in months] +
    [f"max_{m}" for m in months] +
    [f"min_{m}" for m in months]
)
df_temp = df_temp[ordered_cols]


df_temp.head()
```

# Statistical Analysis

```python
df_final.columns
```

```python
continuous=['seclusion','urban','wellness','cuisine','nightlife','beaches','nature','adventure','culture','longitude','latitude','avg_1', 'avg_2', 'avg_3', 'avg_4',
       'avg_5', 'avg_6', 'avg_7', 'avg_8', 'avg_9', 'avg_10', 'avg_11',
       'avg_12', 'max_1', 'max_2', 'max_3', 'max_4', 'max_5', 'max_6', 'max_7',
       'max_8', 'max_9', 'max_10', 'max_11', 'max_12', 'min_1', 'min_2',
       'min_3', 'min_4', 'min_5', 'min_6', 'min_7', 'min_8', 'min_9', 'min_10',
       'min_11', 'min_12']
categories=['city','country','region','culture','ideal_durations','Day trip', 'Long trip', 'One week', 'Short trip',
       'Weekend']
```

```python
from scipy import stats
```

```python
#TWO SAMPLE T - TEST : continuous vs continuous
def two_sample(d1,d2):
  t=0
  f=0
  for i in  range(31):
    sample1=d1.sample(frac=0.03)
    sample2=d2.sample(frac=0.03)
    t_test,p_value=stats.ttest_ind(sample1,sample2)
    if p_value < 0.05:
      f=f+1
    else:
      t=t+1
  if t>f:
    return True
  else:
    return False
```

```python
#CHI-SQUARE TEST - category vs category
def chisquare_cat_vs_cat(d1,d2):
    return True if stats.chi2_contingency(pd.crosstab(d1,d2))[1] < 0.05 else False
```

```python
#ANNOVA TEST - continious vs category
def annova_test(d1,d2):
    group= df[d2].unique()
    data = {}
    for i in group:
        data[i]=df[d1][df[d2]==i]
    f_value,p_value=stats.f_oneway(*[i for i in data.values()])
    return False if p_value < 0.05 else True
```

```python
#All the statistical test being done across all the columns
final = {}
for i in df.columns:
    final[i]={}
    for j in df.columns:
        if (i in continuous) and (j in continuous):
            result=two_sample(df[i],df[j])
        elif (i in continuous) and (j in categories):
            result=annova_test(i,j)
        elif (i in categories) and (j in continuous):
            result=annova_test(j,i)
        elif (i in categories) and (j in categories):
            result=chisquare_cat_vs_cat(df[i],df[j])
        if result:
            final[i][j]=1
        else:
            final[i][j]=0
```

```python
#The result of the statistical test which was in a dictionary is put into a dataframe
final_df = pd.DataFrame(final)
final_df
```
